```python
# Matplotlib:
#  Useful for making plots
```

```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for out plots
import numpy as np
```

```python
x= np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# ploting the data
```

```python
# sin wave

plt.plot(x,y)
plt.show()
```

```python
# cos wave

plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('angel')
plt.ylabel('sine wave')
```

```python
# parabola

x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'b+')
plt.show()
# b shows the color and symbol show the signs
```

```python
plt.plot(x,y,'rx')
```

```python
#
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x),'r--')
plt.show()
```

```python
# bar plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'Hindi', 'Spanish', 'German', 'Latin']
people = [100,50,150,40,70]
ax.bar(languages, people)
plt.xlabel('Language')
plt.ylabel('People')
plt.show()
```

```python
# pie chart

fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'Hindi', 'Spanish', 'German', 'Latin']
people = [100,50,150,40,70]
ax.pie(people,labels=languages, autopct='%1.1f%%' )
plt.show()
```

```python
# scatter plot 
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
# build 3d satter plot

fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, c=z, cmap='Blues')
plt.show()
```
