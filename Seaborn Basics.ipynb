```python
# house price data
from sklearn.datasets import fetch_openml
house_boston = fetch_openml(name='boston') 
house = pd.DataFrame(house_boston.data, columns= house_boston.feature_names)
house['price'] = house_boston.target
house.head()
```

```python
print(house_boston)
```

```python
# Distribution Plot
```

```python
sns.displot(house['price'])
```

```python
sns.distplot(house['price'])
```

```python
# Corelation
#    1. Positive Corelation
#    2. Negative Corelation
```

```python
# Heat Map
```

```python
corelation = house.corr()
```

```python
# constructiong a heat map
plt.figure(figsize=(10,10))
sns.heatmap(corelation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
