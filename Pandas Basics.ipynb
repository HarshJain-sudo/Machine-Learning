```python
# import pandas library
import pandas as pd
import numpy as np
```

```python
# create a Pandas DataFrame
```

```python
mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2],
    'price' :[10000,200,3000]
}

data_frame = pd.DataFrame(mydataset)
print(mydataset)
```

```python
# Pandas Series

a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)

a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)

# Key/Value Objects as Series

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
```

```python
# DataFrames

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

myvar = pd.DataFrame(data)

print(myvar)
```
